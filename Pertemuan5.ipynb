##Library

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics.cluster import fowlkes_mallows_score
```

#Latihan 1

##Dataset

```python
# data tabular
df = pd.DataFrame(
    {
        "ID Buah": pd.Series([1,2,3,4,5]),
        "Diameter": pd.Series([7.0,7.1,6.5,6.6,6.8]),
        "Berat": pd.Series([165,170,180,195,200])
    }
)
```

##Pre-processing

```python
# pemilihan feature
X = df.drop("ID Buah", axis=1)
X
```

##Modeling menggunakan Agglomerative Clustering

```python
clustering = AgglomerativeClustering().fit(X)
df["cluster"] = clustering.labels_
df
```

#Latihan2

```python
df_raw = pd.read_csv("https://raw.githubusercontent.com/rahmantaufik27/dataset/main/study/student_score_pass.csv")
df_raw
```

Preprocessing

```python
# feature selection
x = [0 if x == 'not pass' else 1 for x in df_raw.dropna()['pass_test']]
df_training = df_raw.drop(['npm', 'pass_test'], axis = 1)
# hapus NaN
df_training = df_training.dropna()
df_training
```

Modeling

```python
model = AgglomerativeClustering().fit(df_training)
# hasil clustering
df_training["kluster"] = model.labels_
df_training
```

```python
len(df_training["kluster"].unique())
```

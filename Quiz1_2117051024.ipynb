```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Akses data

```python
df = pd.read_csv('https://raw.githubusercontent.com/adivii/datasets/main/Artificial%20Intelligence/Pertemuan%205/Wine_Quality_Data.csv')
df
```

Preprocessing

```python
# Feature Selection
y = df['quality']
df = df.drop('quality', axis=1)
df = df.drop('Id', axis=1)

# Hapus NaN
df = df.dropna()

df
```

```python
# Split data
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=20)
```

Modeling

```python
# Model klasifikasi linear menggunakan Logistic Regression
model = LogisticRegression()

# Buat model train set
model.fit(x_train, y_train)
```

```python
# Prediksi menggunakan test set
pred = model.predict(x_test)
```

```python
# Menghitung akurasi
accuracy = accuracy_score(y_test, pred)

print('Accuracy: ', accuracy)
```

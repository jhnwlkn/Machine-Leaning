```python
import pandas as pd
import numpy as np

from mlxtend.frequent_patterns import apriori, association_rules
```

# Dataset

* Membuat dataset transaksi sebagai dataframe
* Cek dataset transaksi

```python
# Membuat data  tabular yang terdiri dari data transaksi
df = pd.DataFrame({'ID Transaksi' : pd.Series([1,2,3,4,5]),
                   'Item' : pd.Series([['Susu', 'Roti', 'Biskuit', 'Mentega'],
                                       ['Susu', 'Biskuit', 'Gula', 'Kopi'],
                                       ['Gula', 'Kopi', 'Susu'],
                                       ['Kopi', 'Biskuit', 'Susu', 'Gula'],
                                       ['Mentega', 'Roti', 'Kopi']])
                   }
                  )

# Menampilkan info dataframe
df.info()
```

```python
df
```

# Encoding Data Transaksi

```python
# encoding  = memecah data
df = df.explode('Item')
df
```

```python
basket = (df.groupby(['ID Transaksi', 'Item'])['Item'].count()\
                                      .unstack().reset_index().fillna(0) \
                                      .set_index('ID Transaksi'))
basket.head()
```

#Algoritma Apriori

Eksekusi program apriori menggunakan library mlxtend dengan fungsi apriori

# Apriori Tahap 1

Eliminasi itemset berdasarkan minimun support

```python
frequent_itemsets = apriori(basket, min_support = 0.6, use_colnames=True)

# print(list(frequent_itemsets))

frequent_itemsets
```

#Apriori Tahap 2
Perhitungan confidence

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)
rules[['antecedents', 'consequents', 'support', 'confidence']]
```

Manual

```python
transactions = [
    ['Susu', 'Roti', 'Biskuit', 'Mentega'],
    ['Susu', 'Biskuit', 'Gula', 'Kopi'],
    ['Gula', 'Kopi', 'Susu'],
    ['Kopi', 'Biskuit', 'Susu', 'Gula'],
    ['Mentega', 'Roti', 'Kopi']
]

item_counts = {}
for transaction in transactions:
    for item in transaction:
        if item not in item_counts:
            item_counts[item] = 1
        else:
            item_counts[item] += 1

min_support = 0.4
frequent_items = {}
for item, count in item_counts.items():
    if count / len(transactions) >= min_support:
        frequent_items[item] = count / len(transactions)

candidates = []
for item1 in frequent_items.keys():
    for item2 in frequent_items.keys():
        if item1 != item2:
            candidates.append([item1, item2])

candidate_counts = {}
for transaction in transactions:
    for candidate in candidates:
        if set(candidate).issubset(set(transaction)):
            candidate_key = ','.join(candidate)
            if candidate_key not in candidate_counts:
                candidate_counts[candidate_key] = 1
            else:
                candidate_counts[candidate_key] += 1

frequent_itemsets = {}
for candidate, count in candidate_counts.items():
    support = count / len(transactions)
    if support >= min_support:
        frequent_itemsets[candidate] = support

print("Frequent Itemsets:")
for itemset, support in frequent_itemsets.items():
    print(itemset.split(','), support)
```

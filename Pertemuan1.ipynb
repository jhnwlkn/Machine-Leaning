# Conditional

Mengeksekusi nilai berdasarkan kondisi

```python
if 7 == A:
  print('Nilai 7 sama dengan nilai A yang mana adalah', A)
else:
  print('Nilai 7 tidak sama dengan nilai A yang mana adalah', A)
```

```python
if 70 in nilai:
  print('ada nilai 70 di dalam list')
else:
  print('tidak ada nilai di dalam list')
```

# Library

Sekumpulan kode yang memiliki fungsi-fungsi teretentu dan dapat dipanggil kedalam program lain

```python
# deklarasi library dengan fungsi yang ada didalamnya
import numpy as np

list1 = [1,2,3,4,5]
vtr1 = np.array(list1)
scalarValue = 10

# operasi perkalian vector scalar
vtrScalar = vtr1 * scalarValue
print(vtrScalar)
```

```python
import pandas

mhs = ['Andi', 'Budi', 'Caca', 'Didi']
scr = [7,6,10,8]
ket = []

for i in scr:
  if i < 7:
    ket.append('belum lulus')
  else:
    ket.append('lulus')


data = {  'mahasiswa' :  mhs,
          'nilai' : scr,
          'keterangan' : ket
}


df = pandas.DataFrame(data)
df
```

```python
#make sure have run answer cell number 2

import pandas

ps = df['keterangan'].value_counts(normalize=True)
psFr = ps.apply(lambda x : '{:.2%}'.format(x))

dfPs = pandas.DataFrame({'nilai' : df['keterangan'].value_counts(), 'Persentase' : psFr})

print(df)
print(dfPs)
```
